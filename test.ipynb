```python
import multitest
```

```python
VaRtest = multitest.VaR(tickers=['BTG', 'CET', 'GORO', 'NGD', 'UAMY', 'ASM', 'PLG', 'UEC', 'AAPL', 'LODE', 'USAS', 'MUX', 'XPL'], start = "2010-01-01", end = "2018-12-31", VaR_level=0.05, window=500)
```

```python
VaRtest.calculate_VaR()
```

```python
VaRtest.VaR_forecast.plot()
```

```python
VaRtest.VaR_forecast[VaRtest.VaR_forecast['VaR']>VaRtest.VaR_forecast['Portfolio']].count()['VaR']
```
